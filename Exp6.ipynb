```python
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv('Iris.csv')
print(data)
```

```python
data['Species'].unique()
```

```python
data.drop(columns = 'Id',inplace=True)
```

```python
g = sns.relplot(x = "SepalLengthCm",y = "SepalWidthCm",data=data,hue = 'Species')
```

```python
data.plot.area(y=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],alpha=0.4,figsize=(12, 6));
```

```python
x = data.iloc[:,0:4].values
y = data.iloc[:,4].values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state=0)
gb = GaussianNB()
gb.fit(x_train,y_train)
y_pred = gb.predict(x_test)
```

```python
cm = metrics.confusion_matrix(y_test,y_pred)
print(cm)
print("Accuracy",metrics.accuracy_score(y_pred,y_test))
print("Precision",metrics.precision_score(y_pred,y_test,average='micro'))
print("Recall",metrics.recall_score(y_pred,y_test,average='micro'))
print("F1",metrics.f1_score(y_pred,y_test,average='micro'))
```
